```python
# Q1. Given a list of product prices, write a Python program to filter out all products priced below $50, then apply a 10% discount to the remaining products using filter(), map(), and lambda functions.
prices = [30, 55, 100, 45, 75, 20, 120]
filtered_prices = list(filter(lambda x: x >= 50, prices))
discounted_prices = list(map(lambda x: x * 0.9, filtered_prices))

print("Original prices:", prices)
print("Filtered (>= $50):", filtered_prices)
print("After 10% discount:", discounted_prices)
```

```python
#Q2. Given a list of employee work hours logged in a week, write a Python program to filter out employees who worked less than 40 hours and then convert the remaining hours to overtime hours calculated as (hours - 40) using filter(), map(), and lambda.

work_hours = [35, 40, 45, 38, 50, 60, 25]
eligible = list(filter(lambda h: h >= 40, work_hours))
overtime = list(map(lambda h: h - 40, eligible))

print("Original hours:", work_hours)
print("Worked 40+ hours:", eligible)
print("Overtime hours:", overtime)
```

```python
#Q3. Given a list of customer feedback strings, write a Python program to filter out feedback messages shorter than 20 characters, then convert the remaining feedback messages to lowercase using filter(), map(), and lambda.

feedback = [
    "Great service!",
    "The product quality was amazing.",
    "Too expensive.",
    "Loved it!",
    "Customer support was excellent and quick."
]
long_feedback = list(filter(lambda f: len(f) >= 20, feedback))
lowercase_feedback = list(map(lambda f: f.lower(), long_feedback))

print("Original feedbacks:", feedback)
print("Filtered (>= 20 chars):", long_feedback)
print("Lowercase feedbacks:", lowercase_feedback)
```

```python
#Q4. Given a dictionary of employee names and their salaries, write a Python program to filter out employees with salaries less than $60,000 and then apply a 5% raise to the remaining salaries using filter() and map() with lambda.

salaries = {
    "John": 55000,
    "Sara": 60000,
    "Mike": 75000,
    "Emma": 58000,
    "Alex": 90000
}

filtered = dict(filter(lambda item: item[1] >= 60000, salaries.items()))

raised = dict(map(lambda item: (item[0], item[1] * 1.05), filtered.items()))

print("Original salaries:", salaries)
print("Filtered (>= 60,000):", filtered)
print("After 5% raise:", raised)
```

```python
# Q5. Given a list of exam scores, write a Python program to filter out all below passing score (e.g., less than 50), then map the remaining scores to a letter grade according to the scale: 90+ = 'A', 80–89 = 'B', 70–79 = 'C', 60–69 = 'D', 50–59 = 'E'.
scores = [45, 67, 82, 91, 55, 73, 49, 88]
passed = list(filter(lambda s: s >= 50, scores))
grades = list(map(lambda s: 
                  'A' if s >= 90 else
                  'B' if s >= 80 else
                  'C' if s >= 70 else
                  'D' if s >= 60 else
                  'E', passed))

print("Original scores:", scores)
print("Passed scores:", passed)
print("Letter grades:", grades)
```
